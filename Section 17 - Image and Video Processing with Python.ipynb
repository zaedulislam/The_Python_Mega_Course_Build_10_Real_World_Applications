# 138. Loading, Displaying, Resizing, and Creating Images

```python
import cv2
# Google colab crashes if we try to display image using cv2.imshow() 
# instead "import from google.colab.patches import cv2_imshow" and display using cv2_imshow(<image>)
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/The Python Mega Course 2022: Build 10 Real-World Programs/galaxy.jpg", 0)
print("Type of the image object: ", type(image))
print(image)
print("Shape of the image: ", image.shape)
print("Dimension of the numpy array: ", image.ndim)
```

### Display the Image

```python
cv2.imshow("Galaxy", image)
```

```python
cv2_imshow(image)
# The window will stay for 2000 ms = 2s and then it closed
cv2.waitKey(2000)
cv2.destroyAllWindows()
```

```python
resizedImage = cv2.resize(image, (1000, 500))
cv2_imshow(resizedImage)
# The window will stay for 2000 ms = 2s and then it closed
cv2.waitKey(2000)
cv2.destroyAllWindows()
```

```python
resizedImage = cv2.resize(image, (int(image.shape[1]/2), int((image.shape[0]/2))))
print("Shape of the resized image: ", resizedImage.shape)
cv2_imshow(resizedImage)
# Saving the resized image
cv2.imwrite("Galaxy Resized.jpg", resizedImage)
# The window will stay for 2000 ms = 2s and then it closed
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
imageRGB = cv2.imread("/content/drive/MyDrive/Colab Notebooks/The Python Mega Course 2022: Build 10 Real-World Programs/galaxy.jpg", 1)
print("Type of the image object: ", type(imageRGB))
print(imageRGB)
print("Shape of the image: ", imageRGB.shape)
print("Dimension of the numpy array: ", imageRGB.ndim)
```
